### Step 3: Handle Missing Values

#### •	For numeric columns (e.g., Quantity, UnitPrice): Replace missing values with mean or median.

```python
p1["Quantity"] = p1["Quantity"].fillna(p1["Quantity"].median())
p2["Quantity"] = p2["Quantity"].fillna(p2["Quantity"].median())

p1["UnitPrice"] = p1["UnitPrice"].fillna(p1["UnitPrice"].mean())
p2["UnitPrice"] = p2["UnitPrice"].fillna(p2["UnitPrice"].mean())

p1["SalesAmount"] = p1["SalesAmount"].fillna(p1["SalesAmount"].median())
p2["SalesAmount"] = p2["SalesAmount"].fillna(p2["SalesAmount"].median())
```
